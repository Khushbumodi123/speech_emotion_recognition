```python
import os
import librosa
import librosa.display
import numpy as np
import pandas as pd
import IPython.display as ipd
import matplotlib.pyplot as plt
```

 Loading Dataset

```python
path = glob.glob('/content/drive/MyDrive/speech_emotion_recognition/*.*')
audio_df = pd.read_csv(path[1])
audio_df.head()
```

# Exploratory Data Analysis

```python
barWidth = 0.25
fig = plt.figure(figsize =(12, 8))

X = ['neutral','calm','happy','sad','angry','fearful','disgust','surprised']

Yclass1 = [(audio_df[(audio_df.emotion == 'neutral')].gender =='male').sum(),
           (audio_df[(audio_df.emotion == 'calm')].gender =='male').sum(),
           (audio_df[(audio_df.emotion == 'happy')].gender =='male').sum(),
           (audio_df[(audio_df.emotion == 'sad')].gender =='male').sum(),
           (audio_df[(audio_df.emotion == 'angry')].gender =='male').sum(),
           (audio_df[(audio_df.emotion == 'fearful')].gender =='male').sum(),
           (audio_df[(audio_df.emotion == 'disgust')].gender =='male').sum(),
           (audio_df[(audio_df.emotion == 'surprised')].gender =='male').sum()]

Yclass2 = [(audio_df[(audio_df.emotion == 'neutral')].gender =='female').sum(),
           (audio_df[(audio_df.emotion == 'calm')].gender =='female').sum(),
           (audio_df[(audio_df.emotion == 'happy')].gender =='female').sum(),
           (audio_df[(audio_df.emotion == 'sad')].gender =='female').sum(),
           (audio_df[(audio_df.emotion == 'angry')].gender =='female').sum(),
           (audio_df[(audio_df.emotion == 'fearful')].gender =='female').sum(),
           (audio_df[(audio_df.emotion == 'disgust')].gender =='female').sum(),
           (audio_df[(audio_df.emotion == 'surprised')].gender =='female').sum()] 

X_axis1 = np.arange(len(X))
X_axis2 = [x + barWidth for x in X_axis1]

plt.grid(axis = 'y')
plt.bar(X_axis1, Yclass1, width = barWidth, label = 'Male')
plt.bar(X_axis2, Yclass2, width = barWidth, label = 'Female')
plt.legend()
plt.xticks(X_axis1+0.1, X)
plt.xlabel("Emotion", fontsize=16)
plt.ylabel("Count", fontsize=16)
plt.title("Emotions value counts", fontsize=20)
plt.legend()
plt.show()
path_save=glob.glob('/content/drive/MyDrive/speech_emotion_recognition/EDA_images/')
fig.savefig(path_save[0] + 'Emotions_value_counts.jpg', bbox_inches='tight', dpi=150)
```

# Male Neutral

```python

path = glob.glob('/content/drive/MyDrive/speech_emotion_recognition/data/Actor_01/*.*')
audio = path[0]
data, sample_rate = librosa.load(audio)

# DISPLAY WAVEPLOT
plt.figure(figsize=(8, 4))
librosa.display.waveplot(data, sr=sample_rate)
plt.title('Waveplot - Male Neutral')
path_save=glob.glob('/content/drive/MyDrive/speech_emotion_recognition/EDA_images/')
plt.savefig(path_save[0] + 'Waveplot_MaleNeutral.jpg')

#DISPLAY SPECTOGRAM
plt.figure(figsize=(8, 4))
spectrogram = librosa.feature.melspectrogram(y=data, sr=sample_rate, n_mels=128,fmax=8000) 
spectrogram = librosa.power_to_db(spectrogram)
librosa.display.specshow(spectrogram, y_axis='mel', fmax=8000, x_axis='time');
plt.title('Mel Spectrogram - Male Neutal')
plt.colorbar(format='%+2.0f dB')

path_save=glob.glob('/content/drive/MyDrive/speech_emotion_recognition/EDA_images/')
plt.savefig(path_save[0] + 'Spectrogram_MaleNeutral.jpg')

ipd.Audio(audio)
```
